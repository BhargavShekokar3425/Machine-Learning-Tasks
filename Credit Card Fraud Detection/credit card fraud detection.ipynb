This is a Credit Card Fraud prediction system

**Work Flow**

Credit Card data -> Data pre processing -> Data analysis -> Train test split -> Logistic Regression Model

Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#load the dataset in DF
credit_data = pd.read_csv('creditcard.csv')
```

```python
credit_data.describe()
```

```python
credit_data['Class'].value_counts()
```

```python
credit_data.isnull().sum()
```

This dataset is highly unbalanced

0--> Normal Transaction

1--> Fraudulent transaction

```python
#separating the data for analysis
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape,fraud.shape)
```

```python
#statisctical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_data.groupby('Class').mean()
```

Under sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transaactions

Number of Fraudulent Transactions -> 492

We take random 492 data points from legit dataset and join it with fraud dataset to make it a good dataset for training

```python
legit_sample =legit.sample(n=492)
```

Concatenating two Dataframes

```python
new_data =pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.shape
```

```python
new_data['Class'].value_counts()
```

Splitting of data

```python
X = new_data.drop(columns='Class',axis=1)
Y = new_data['Class']
```

```python
X
```

```python
Y
```

Train -test split

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,stratify=Y,random_state=1,test_size=0.2)
```

```python
X_train,X_test,Y_train,Y_test
```

Training your model : Logistic Regression

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train,Y_train)
```

Model Analyisis

```python
#accuracy on training data
X_train_prediction = classifier.predict(X_train)
X_train_accuracy = accuracy_score(Y_train,X_train_prediction)
print("The accuracy score for training dataset is ",X_train_accuracy)


#accuracy on testing data
X_test_prediction = classifier.predict(X_test)
X_test_accuracy = accuracy_score(Y_test,X_test_prediction)
print("The accuracy score for testing dataset is ",X_test_accuracy)
```
