This is Parkinson Disease Detector using svm(Support Vector Machine)

Importing Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
# to normalize the data
```

```python
dataset = pd.read_csv("Parkinsson disease.csv")
```

```python
dataset.head()
```

Data Analysis

```python
dataset['status'].value_counts()
```

147 pateints and 48 healthy people

```python
#Seperating features and Target
X = dataset.drop(columns=['name','status'],axis=1)
Y = dataset['status']
```

```python
X,Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,stratify=Y,random_state=1,test_size=0.2)
```

Data Standardization

```python
scalar = StandardScaler()
```

```python
scalar.fit(X_train)
```

```python
X_train = scalar.transform(X_train)

X_test = scalar.transform(X_test)
```

```python
X_train
```

```python
X_test
```

Model Training: Support Vector Machine Model

```python
model =svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

Model Analysis

```python
X_train_prediction = model.predict(X_train)
X_train_accuracy = accuracy_score(Y_train,X_train_prediction)
print(" The accuracy of training data is ",X_train_accuracy)

X_test_prediction = model.predict(X_test)
X_test_accuracy = accuracy_score(Y_test,X_test_prediction)
print(" The accuracy of testing data is ",X_test_accuracy)
```

Prediction System

```python
input_data=(116.676,137.871,111.366,0.00997,0.00009,0.00502,0.00698,0.01505,0.05492,0.517,0.02924,0.04005,0.03772,0.08771,0.01353,20.644,0.434969,0.819235,-4.117501,0.334147,2.405554,0.368975)

#changing to numpy array
input_data_np = np.asanyarray(input_data)
#reshaping array
input_data_reshape = input_data_np.reshape(1,-1)

#standardise the data
std_data = scalar.transform(input_data_reshape)

prediction = model.predict(std_data)
print(" the prediction is :",prediction[0])

if(prediction[0]==1):
    print("The person has parkinson disease")
else:
    print("The person is healthy")


```
